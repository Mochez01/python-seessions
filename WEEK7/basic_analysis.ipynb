```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    # Load genguine dataset
    df = pd.read_csv("penguins_size.csv")
    
    # Display first few rows
    print("First 5 rows of the dataset:")
    print(df.head())

    # Check data types and missing values
    print("\nData Types:")
    print(df.dtypes)

    print("\nMissing Values:")
    print(df.isnull().sum())

    # Clean dataset 
    df_cleaned = df.dropna()

except FileNotFoundError:
    print("Error: File not found.")
except Exception as e:
    print("An error occurred:", e)
```

```python
print(df_cleaned.describe())
```

```python
# Group by species and calculate mean of flipper length
grouped = df_cleaned.groupby('species')['flipper_length_mm'].mean()
print(grouped)
```

Gentoo species have the longest flipper length of 217.23 mm

```python
# 2. Bar Chart: Average body mass per species
plt.figure(figsize=(8, 5))
df.groupby('species')['body_mass_g'].mean().plot(kind='bar')
plt.title("Bar Chart: Average body mass (g) by Species")
plt.xlabel("Species")
plt.ylabel("body Mass (g)")
plt.tight_layout()
plt.show()
```

```python
# Histogram: Distribution of flipper length
plt.figure(figsize=(8, 5))
plt.hist(df_cleaned['flipper_length_mm'], bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram: flipper length Distribution")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python

# 4. Scatter Plot: Flipper Length vs Body Mass
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    data=df_cleaned,
)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.tight_layout()
plt.show()
```
